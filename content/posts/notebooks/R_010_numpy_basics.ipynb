```python
import numpy as np
import matplotlib.pyplot as plt

print("numpy:", np.__version__)
```

```python
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float64)

print("a:", a, "shape=", a.shape, "dtype=", a.dtype)
print("b:\n", b, "\nshape=", b.shape, "dtype=", b.dtype)

zeros = np.zeros((2, 3))
ones = np.ones((2, 3))
ar = np.arange(0, 10, 2)

print("zeros:\n", zeros)
print("ones:\n", ones)
print("arange:", ar)
```

```python
x = np.arange(1, 13).reshape(3, 4)
print("x:\n", x)

print("x[0, :]:", x[0, :])
print("x[:, 2]:", x[:, 2])
print("x[1:, 1:3]:\n", x[1:, 1:3])
```

```python
y = np.arange(1, 7)
y2 = y.reshape(2, 3)
yt = y2.T

print("y:", y)
print("y2:\n", y2)
print("y2.T:\n", yt)
```

```python
m = np.arange(1, 13).reshape(3, 4)
v = np.array([1, 10, 100, 1000])

out = m + v
print("m:\n", m)
print("v:", v)
print("m + v:\n", out)
```

```python
t = np.linspace(0, 2*np.pi, 200)
s = np.sin(t)

plt.figure()
plt.plot(t, s)
plt.title("NumPy + Matplotlib sanity")
plt.xlabel("t")
plt.ylabel("sin(t)")
plt.show()
```
